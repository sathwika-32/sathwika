```python
plt.figure(figsize=(12,8))
sns.heatmap(round(sgemm_df[[columns for columns in sgemm_df.describe().columns if columns not in dependent_var]].corr().abs(),3),annot=True,cmap='coolwarm')
```

```python
X=sgemm_df.drop([columns for columns in dependent_var],axis=1)
y=sgemm_df['Run2 (ms)']
```

```python
plt.figure(figsize=(6,4))
sns.histplot(y,color='red',kde=True)
plt.show()
```

As we can see our target is not looking normally distributed let's take log on y variable and draw the plot.

```python
plt.figure(figsize=(6,4))
sns.histplot(np.log(y),color='red',kde=True)
plt.show()
```

```python
y_final=np.log(y)
```

Split-train-test data

```python
# import libraray
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Splitting the dataset into the Training set and Test set
X_train, X_test, y_train, y_test = train_test_split(X, y_final, test_size = 0.25, random_state = 0)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

Linear Regression

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
regressor.intercept_
```

```python
regressor.coef_
```

```python
y_pred_train=regressor.predict(X_train)
```

```python
y_pred=regressor.predict(X_test)
```

Evaluation Matrix


```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,accuracy_score
```

```python
r2_score(y_test, y_pred)
```

```python
print("Adjusted R2 : ",1-(1-r2_score((y_test), (y_pred)))*((X_test.shape[0]-1)/(X_test.shape[0]-X_test.shape[1]-1)))
```

```python
plt.scatter(y_test, y_pred, color='blue', label='Data Points')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', label='Regression Line')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Linear Regression: Actual vs Predicted')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.plot(y_pred)
plt.plot(np.array(y_test))
plt.legend(["Predicted","Actual"])
plt.xlabel('No of Test Data')
plt.show()
```
